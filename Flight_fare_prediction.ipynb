**Exploratory Data Analysis (EDA)**

1. from the description we can see that Date_of_Journey is a object data type, Therefore, we have to convert this to timestamp so as to use this column for prediction.

2. for this we require pandas to_determine to convert object data type to datetime dtype.
*.dt.day method will extract only day of that date*

*.dt.month method will extract only month of that date*

**Creating two features Journey_Day and Journey_Month**

```python
train_data["Journey_day"] = pd.to_datetime(train_data.Date_of_Journey, format= "%d/%m/%Y").dt.day
```

```python
train_data["Journey_month"] = pd.to_datetime(train_data.Date_of_Journey, format = "%d/%m/%Y").dt.month
```

```python
train_data.head()
```

**axis = 1 refers that drop coloumn**

```python
# Since we have converted Date_of_Journey column into integers, Now we can drop it as it is of no use.

train_data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
# Departure time is when a plane leaves the gate
# Similarly to Date_of_Journey we can extract values from Dep_Time

# Extracting Hours
train_data["Dep_hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour

#Extracting Minutes
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute

# Now we can drop Dep_Time as it is of no use
train_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# Arrival Time is when the plane pulls up to the gate.
#Similar to the Date_of_Journey we can extract values from Arrival time:

# Extracting Hours:
train_data["Arrival_hour"] = pd.to_datetime(train_data["Arrival_Time"]).dt.hour

# Extracting Minutes:
train_data["Arrival_min"] = pd.to_datetime(train_data["Arrival_Time"]).dt.minute

#dropping the arrival TIme:
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# Time taken by plane to reach destination is duration
# it is the difference between Departure Time and Arrival Time

# Assigning and converting Duration Column into list
duration = list(train_data["Duration"])


# what we are doing below:
# We going through every element in duration list
# for every element we are checking if string contains both hour and min or just min or just hour
# we are adding 0 hour or 0 minute accordingly because we want uniformity in the data


for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"    # Adds 0 minutes
        else:
            duration[i] = "0h " + duration[i]        # Adds 0 hour


# Now making the duration hour and duration min list
# we will then make two new features and assign these list to the features

duration_hours = []
duration_mins = []

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0])) # Extracting hours
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1])) # Extracting Minutes
```

```python
# adding features duration_hours ans duration_mins to the train_data dataframe

train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data.head()
```

```python
train_data.drop(["Duration"], axis = 1, inplace = True)
```

**Hyperparameter Tuning**

Choose following method for hyperparameter tuning

    1. RandomizedSearchCV --> Fast
    2. GridSearchCV
    
Assign hyperparameters in form of dictionery

Fit the model

Check best paramters and best score

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
plt.figure(figsize = (8,8))
sns.distplot(y_test-prediction)
plt.show()
```

```python

plt.figure(figsize = (8,8))
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```
